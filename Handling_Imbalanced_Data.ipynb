Imbalanced dataset:

a dataset with unequal class distribution


in this example we will use an inbalaced dataset to make it balanced

dataset link :
https://drive.google.com/file/d/1WkcjUn5tVJDUtkN6d2_gztiT1FmIIVDP/view

```python
#importing needed libraries
import pandas as pd
import numpy as np
```

```python
#loading the dataset
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
#distribution of data in the 2 classes
credit_card_data['Class'].value_counts()
```

as we see the data is imbalaced

1----> fraudlent transactions

0----> normal/legit transactions

```python
#separiting the legit and fraudlent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

under-sampling

we will build a sample dataset containinig similar distribution of legit and fraudlent transcations


the number of legit transactions is 492

```python
legit_sample = legit.sample(n=492)
print(legit_sample.shape)
```

```python
new_dataset = pd.concat([legit_sample, fraud] , axis= 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

as we see we balanced the data (we can take equal numbers ) but i decided to make the values closer to each other
