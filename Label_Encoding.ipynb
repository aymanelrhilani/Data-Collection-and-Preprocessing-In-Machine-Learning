what is label encoding ?

--> label encoding is converting the labels into numeric form

in this example we will use 2 datasets , iris_data.csv and data.csv

```python
#importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
cancer_data = pd.read_csv('/content/data.csv')
```

```python
# lets print the 100 first rows
cancer_data.head(100)
```

as we see the column called diagnosis has a value "M" and "B",the concept of label encoding is giving this alphabets a numerical value in order to make it easy to process.

```python
#finding the count of diffrent labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the encoder function to a variable
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis) # here we encoded the two values M and B to 1 and 0 and we put them in variable called labels
```

```python
# appending the labels to the dataframe
cancer_data['target']= labels
```

```python
print(cancer_data)
```

as we see new column called target has showed up

M (malignant) --->1

B(benign)--->0

---



```python
cancer_data['target'].value_counts() #the same results as diagnosis
```

```python
# this how can we delete a column from dataframe
del cancer_data['diagnosis']
```

```python
#checking if it is deleted
print(cancer_data.head())
```
